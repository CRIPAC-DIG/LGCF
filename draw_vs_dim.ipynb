```python
import matplotlib.pyplot as plt
```

```python
light_recall_20 = [0.0840, 0.1081, 0.1172, 0.1309, 0.1368]
light_ndcg_20 = [0.0500, 0.0664, 0.0743, 0.0826, 0.0863]

hgcf_recall_20 = [0.0839, 0.1181, 0.1313, 0.1405, 0.1455]
hgcf_ndcg_20 = [0.0488, 0.0726, 0.0820, 0.0889, 0.0909]

lgcf_recall_20 = [0.0961, 0.1302, 0.1406, 0.1468, 0.1503]
lgcf_ndcg_20 = [0.0568, 0.0792, 0.0868, 0.0916, 0.0945]
```

```python
xs = [10, 20, 30, 40, 50]
```

```python
plt.plot(xs, light_recall_20, 'o-', label='LightGCN', linewidth=3)
plt.plot(xs, hgcf_recall_20,  '^-', label='HGCF', linewidth=3)
plt.plot(xs, lgcf_recall_20, 's-', label='LGCF', linewidth=3)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('Dimension', fontsize=12)
plt.ylabel('Recall@20', fontsize=12)
plt.legend(fontsize=12)
plt.savefig('recall.pdf')
# plt.close()
```

```python
plt.plot(xs, light_ndcg_20, 'o-',label='LightGCN', linewidth=3)
plt.plot(xs, hgcf_ndcg_20, '^-',label='HGCF', linewidth=3)
plt.plot(xs, lgcf_ndcg_20, 's-',label='LGCF', linewidth=3)
plt.xlabel('Dimension', fontsize=12)
plt.ylabel('NDCG@20', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.savefig('ndcg.pdf')
# plt.close()
```
