```python
from main import lazy_load_dataset
from argparse import Namespace
import numpy as np
from scipy.stats import powerlaw
import matplotlib.pyplot as plt
import pickle
import random


args = Namespace(**{'dataset':'Amazon-CD'})


data = lazy_load_dataset(args)
user_item = data.user_item_csr
adj = data.adj_train
item_adj = user_item.T @ user_item
user_adj = user_item @ user_item.T
```

```python

degree = np.array(adj.sum(1)).squeeze()



plt.hist(degree, bins=50, density=False, range=(0, 100))
# plt.show()
plt.xlabel("The number of interactions")
plt.ylabel("The number of nodes")
plt.savefig('degree.pdf')
```

```python
with open('adj_4_hop.pkl', 'rb') as f:
    nnei_lst = pickle.load(f) # num of neighbors list
n = len(nnei_lst[0])
sampled_ids = random.sample(range(n), 10)
ks = np.arange(1, len(nnei_lst)+1)
for i in sampled_ids:
    degs = [nnei[i] for nnei in nnei_lst]
    plt.loglog(ks, degs)
plt.rcParams['font.size'] = '13'
plt.xlabel('# hops')
plt.ylabel('# multi-hop neighbors')
plt.savefig('data_ori.pdf')
```
