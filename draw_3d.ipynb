```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
os.mkdir('lorentz_figs')
```

```python
x_range = 0.001
num_samples = 100
x = np.linspace(-x_range, x_range, num_samples)
y = np.linspace(-x_range, x_range, num_samples)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(0.000001 + np.power(X, 2) + np.power(Y, 2))
```

```python

ax = plt.gca(projection='3d')
# colormaps_list = ['viridis', 'plasma', 'inferno', 'magma', 'cividis',
#                 'Greys', 'Purples', 'Blues', 'Greens', 'Reds', 'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'GnBu', 'PuBu', 'YlGnBu','PuBuGn','PuBuGn','BuGn','YlGn','gray','bone','pink','spring','summer','autumn','winter','cool','Wistia','hot','afmhot','gist_heat','copper',
#                 'PiYG','PRGn','BrBG','PuOr','RdGy','RdBu','RdYlBu','RdYlGn','Spectral','coolwarm','bwr','seismic']
colormaps_list = ['OrRd']
for cmp in colormaps_list:
    ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap=plt.get_cmap(cmp))
    # plt.get_cmap('summer')
    ax.grid(False)
    plt.axis('off')
    ax.view_init(elev=360,    # 仰角
                   azim=-120   # 方位角 zheng 往左转
                  ) 
#     plt.show()
    plt.savefig(f'lorentz_figs/{cmp}.pdf')
```
